```python
from bs4 import BeautifulSoup
import requests
import csv
from csv import writer

import pickle
import sys
import re
import pandas as pd
```

```python
url= "http://reports.ieso.ca/public/Adequacy2/"
url_parent= "http://reports.ieso.ca/public/"
```

```python
page = requests.get(url,allow_redirects=True)
soup = BeautifulSoup(page.text)
```

```python
# with open("out_soup3", "w+", encoding='utf-8') as file:
#     file.write(str(soup))
```

```python
xml_cap=soup.find_all("a",href=True)
for element in xml_cap:
    file_name=element.text
    if re.search("\.xml$", file_name) and not re.search("_v", file_name):
        print(element.text)
```

```python
page = requests.get(url_parent,allow_redirects=True)
soup_parent=BeautifulSoup(page.text)
```

```python
dirs=soup_parent.find_all("a",href=True)
for d in dirs:
    file_name=d.text
    if re.search(r"/$", file_name):
        print(d.text)
```
